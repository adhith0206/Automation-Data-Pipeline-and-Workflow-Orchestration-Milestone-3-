## Transform

- **Mengubah kolom data `Smoking` dan `Alcohol_Consumption` menjadi kategorikal**

```python
## Mengubah tipe data `Smoking` dan `Alcohol_Consumption` dari boolean menjadi string

df = df.astype({'Smoking': str, 'Alcohol_Consumption': str})
```

```python
## Mengubah value data di kolom `Smoking` dan `Alcohol_Consumption` dari 0 & 1 menjadi No & Yes

df.replace({'Smoking': {'0': 'No', '1': 'Yes'}, 'Alcohol_Consumption': {'0': 'No', '1': 'Yes'}}, inplace=True)
```

## Data Validation

```python
## Membuat folder data context
context = FileDataContext.create(project_root_dir='./')
```

```python
# Membuat dan memberi nama Datasource
datasource = context.sources.add_or_update_pandas('coffee_health_dataset')

# Membuat dan memberi nama Data Asset
asset = datasource.add_dataframe_asset(name='df')

# Membuat Batch Request
batch_request = asset.build_batch_request(dataframe=df)

# Membuat Expectation Suite
expectation_suite_name = 'expectation_global_coffee_health_dataset'
context.add_or_update_expectation_suite(expectation_suite_name)

# Mbembuat validasi dengan basis expectation suite yang telah dibuat
validator = context.get_validator(
    batch_request = batch_request,
    expectation_suite_name = expectation_suite_name
)

# Melakukan cek validasi
validator.head()
```

### 1. to be unique 

```python
## Melakukan validasi kolom ID harus berbeda tiap baris-nya
validator.expect_column_values_to_be_unique('ID')
```

### 2. to be between min_value and max_value

```python
## Melakukan validasi kolom Age harus minimal 18 tahun dan maksimal 80 tahun, diluar angka tersebut tidak dihitung
validator.expect_column_values_to_be_between(column='Age', min_value=18, max_value=80)
```

### 3. to be in type list

```python
## Melakukan validasi kolom Coffee_Intake harus berupa tipe data angka / float
validator.expect_column_values_to_be_in_type_list(column= 'Coffee_Intake', type_list=['int', 'float'])
```

### 4. to be in set 

```python
## Melakukan validasi pada kolom stress level bahwa nilai kolom hanya berupa low, medium, atau high
validator.expect_column_values_to_be_in_set(column='Stress_Level', value_set=['Low', 'Medium', 'High'])
```

### 5. to not be null

```python
## Melakukan validasi pada kolom ID bahwa tidak ada nilai yang kosong
validator.expect_column_values_to_not_be_null('ID')
```

### 6. mean to be between

```python
## Melakukan validasi nilai mean pada kolom sleep hours berada dalam jangkauan min dan max value
validator.expect_column_mean_to_be_between(column="Sleep_Hours", min_value=3.0, max_value=10.0)
```

### 7. to exist 

```python
## Melakukan validasi pada kolom BMI bahwa kolom tersebut harus ada karena dipakai sebagai acuan dari health issues
validator.expect_column_to_exist('BMI')
```

### Save Expectation Suite

```python
## Save ke Expectation Suite
validator.save_expectation_suite(discard_failed_expectations=False)
```

### Checkpoint

```python
checkpoint_1 = context.add_or_update_checkpoint(
    name = 'checkpoint_1',
    validator = validator,
)
```

```python
## Melakukan checkpoint
checkpoint_result = checkpoint_1.run()
```
